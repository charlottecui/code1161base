# 3. Factors which impact employees leaving

## 3.1 satisfaction level impacts leaving

```python
sns.distplot( data[data.left == 0]['satisfaction_level'], color = 'b')
sns.distplot( data[data.left == 1]['satisfaction_level'], color = 'r')
```

### Average satisfaction levels for stay and leaving 

```python
sl_left_mean = np.mean( data[data.left == 0]['satisfaction_level'] )
sl_left_mean
```

```python
np.std( data[data.left == 0]['satisfaction_level'] )
```

```python
np.mean( data[data.left == 1]['satisfaction_level'] )
```

```python
np.std( data[data.left == 1]['satisfaction_level'] )
```

## 3.2 Salary impacts leaving

As you can see from the plot shows us that around 40% of those with low salaries leave and 25% of those with median salaries leave. It looks like only 10% of those with high salaries leave.There is a big comparison between low, medium, high salary, which shows that lower salary are more likely to leave.

```python
sns.countplot(x='salary', hue='left', data=data)
```

### spread of satisfaction level between the different salary ranges

It shows that low salary has the lowest median satisfaction and the highest spread, and the high salary just the other way around.

```python
sns.boxplot(x='salary', y='satisfaction_level', data=data)
```

### realontionship between salary and sataifaction levels

```python
sns.distplot( data[data.salary == 'low']['satisfaction_level'], color = 'b')
sns.distplot( data[data.salary == 'medium']['satisfaction_level'], color = 'g')
sns.distplot( data[data.salary == 'high']['satisfaction_level'], color = 'r')
```

### How salaries across departments are related to attrition

```python
sns.factorplot( y = "sales",
            col="salary",
            hue = "left",
            data=data,
            kind="count",
            size=7)
```

## 3.3 working accident impacts leaving

```python
work_accident_count = data[['Work_accident', 'left']].groupby(['Work_accident', 'left']).size().reset_index()
work_accident_count.columns = ['Work_accident', 'left', 'count']

sns.factorplot(x="Work_accident", y = 'count', hue="left", data=work_accident_count,
               size=7, kind="bar", palette="muted")
```

```python
work_accident_count = data[['Work_accident', 'left']].groupby(['Work_accident', 'left']).size()
work_accident_percent = work_accident_count.groupby(level=[0]).apply(lambda x: x / x.sum()).reset_index()
```

### leaving percentage of work accident

```python
work_accident_percent.columns = ['Work_accident', 'left', 'percent']
```

```python
sns.factorplot(x="Work_accident", y = 'percent', hue="left", data=work_accident_percent,
               size=7, kind="bar", palette="muted")
```

### How work accidents have impacted the statisfactin level of the empolyees

```python
sns.distplot( data[data.Work_accident == 1]['satisfaction_level'], color = 'r')
sns.distplot( data[data.Work_accident == 0]['satisfaction_level'], color = 'b')
```

It can be seen that large number of people who had lower satisfaction levels, have left the company. Especially, people who have satisfaction level less than 0.5. This makes sense. But there is also a surge in at higher level of satisfaction. Need to understand and deal with these employees with a different stategy.

## 3.4 Number of project impacts leaving

It is very clear that evaluation scores are affected by the number of projects assigned to the employee, also there is a peculiar trend in accounting -- they have a lower last_evaluation score than the other departments at 7 projects.

```python
sns.factorplot(x='number_project', y='last_evaluation', hue='sales', data=data)
```

There is a important relationship between the level of stasfactions and the number of project. Those with high numbers of projects (6 or 7) tend to have extremely low satisfaction levels. Also those with only 2 projects tend to also have lower satisfaction levels.

```python
sns.boxplot(x='number_project', y='satisfaction_level', data=data)
```

## 3.5 Time spend in work impacts leaving

### time spend in company

From the People who have spent 2 years are not leaving the company. But as experience grows people start leaving and highest after they spend 5 years in the company. But once they cross the golden years '7', they are not leaving.

```python
time_spend_count = data[['time_spend_company', 'left']].groupby(['time_spend_company', 'left']).size()
time_spend_percent = time_spend_count.groupby(level=[0]).apply(lambda x: x / x.sum()).reset_index()
time_spend_percent.columns = ['time_spend_company', 'left', 'percent']
```

```python
sns.factorplot(x="time_spend_company", y = 'percent', hue="left", data=time_spend_percent,
               size=7, kind="bar", palette="muted")
```

### time spend in company + salary factor

From the low and medium salaries, shows that leave tend to have spent more time at the company. For those with high salaries, leaving depends on the department. At the high salary level, time spent doesn't vary in accounting for those that left versus those that haven't but it varies pretty wildly for the support and IT departments.

```python
timeplot = sns.factorplot(x='time_spend_company', hue='left', y='sales', row='salary', data=data, aspect=2)
```

## 3.6 evaluation scores impacts leaving

People with low evaluation and very high evaluation are leaving, where as people with average evaluation scores are staying back. 

```python
sns.distplot( data[data.left == 0]['last_evaluation'], color = 'r')
sns.distplot( data[data.left == 1]['last_evaluation'], color = 'g')
```

## 3.7 Promotion impacts leaving 

It is a interesting fact that people who got a promotion in last five years didn't left, but the one that did't get any promotion in last five years more likely left.

```python
pd.crosstab( data.promotion_last_5years, data.left )
```

```python
sns.factorplot(x="promotion_last_5years", hue = 'left', data=data,
               size=7, kind="count", palette="muted")
```

# 4. Conclusion

Overall, there are threee main factors in why employees leave their companies prematurely:

1. A low level of satisfaction whether it be from emotional or work related issues.
2. A low income then what is expected even after promotions or raises.
3. A unsatisfactory amount of work, this could be either too many projects or too less.

The main cause of all these are stress levels and phycological factors that effect their judgement for leaving the company prematurely.
